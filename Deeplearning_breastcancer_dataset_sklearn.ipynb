```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_dataset.keys()
```

```python
breast_cancer_dataset.feature_names
```

```python
#loading to dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.tail()
```

```python
data_frame.shape
```

```python
#info
data_frame.info()
```

```python
data_frame.describe()
```

```python
#target distribution value count
data_frame['label'].value_counts()
```

```python
#average of target
data_frame.groupby('label').mean()
```

```python
#separate feature and target data
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
X.head()
```

```python
#split training data and test data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,random_state=2, test_size=0.2)
```

```python
print(X_train.shape, X_test.shape)
```

```python
## neural network
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up layers of my NN

model = keras.Sequential([
                         keras.layers.Flatten(input_shape=(30,)),
                         keras.layers.Dense(20, activation='relu'),
                         keras.layers.Dense(2, activation='sigmoid')

])
```

```python
#compiling the NN
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

standardize the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
train = model.fit(X_train_std,Y_train,validation_split=0.1,epochs=20)
```

```python
plt.plot(train.history['loss'])
plt.plot(train.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
```

```python
plt.plot(train.history['accuracy'])
plt.plot(train.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
```

Accuracy of the model for test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
